```python
'''
format specifiers = {value: flags} format a value based on what flags are inserted
many decimal places (fixed point)

. (number)f = round to that many decimal places (fixed points)
: (number) = allocate that many spaces
: 03 = allocate and zero pad that many spaces
:< = left justify
:> = right justify
:^ = center align
:+ = use a plus sign to indicate positive value
:= = place sign to leftmost position
: = insert a space before positive numbers
:, = comma separator

'''

price1 = 3.14159
price2 = -9774.67
price3 = 12.34

print(f"price1 is ${price1:.2f}")
print(f"price2 is ${price2:.1f}")
print(f"price3 is ${price3:.3f}")
print(" ")
print(f"price1 is ${price1:.2}")
print(f"price2 is ${price2:.10}")
print(f"price3 is ${price3:.3}")
print(" ")
print(f"pricespace is ${price3: 9}")
print(f"pricezeropadspace is ${price3: 03}")
print(" ")
print(f"price1 is ${price1:<2}")
print(f"price2 is ${price2:>10}")
print(f"price3 is ${price3:^3}")
print(f"price3 is ${price2:+}")
print(f"price3 is ${price2: }")
print(f"price3 is ${price2:,}")
print(f"price3 is ${price2:,.1f}")
print(" ")

```

```python
# While loop - execute code until some condition remain true

name = input("Enter your name: ")

while name == "" :
   print("You did not enter your name")
   name = input("Enter your name: ")

print(f"Hello {name}")

age = int(input("Enter your age: "))

while age == "" or age < 0 or age > 125 :
  print("You did not enter a valid age")
  age = int(input("Enter a valid age: "))

print(f"Your age is {age}")
```

```python
# count down timer project

import time

my_time = int(input("Enter the time in sec: "))

#for x in range(0, my_time):
for x in range(my_time, 0 , -1):
  seconds = x % 60
  #minutes = (x // 60) % 60
  minutes = int(x/60) % 60
  hours = int(x/3600)
  #print(x)
  print(f"{hours:02}:{minutes:02}:{seconds:02}")

  time.sleep(1)

print("Time's up")
```
